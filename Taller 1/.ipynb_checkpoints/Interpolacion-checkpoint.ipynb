# Interpolación


```python
import numpy as np
import cv2
import cmath
from matplotlib import pyplot as plt
from scipy import interpolate
```

## Sin interpolar 
A través de una regla de tres simple enviamos cada pixel de la imagen original a un pixel de la imagen redimencionada. Observamos los espacios sin datos que deben ser rellenados.

```python
def sin_interpolar(img, factor=2): 
    #Control de color o blanco y negro
    if len(img.shape) is 2:
        ans = np.zeros([img.shape[0] * factor, img.shape[1] * factor], dtype=np.uint8)
    else:
        ans = np.zeros([img.shape[0] * factor, img.shape[1] * factor, 3], dtype=np.uint8) #Tamaño de la imagen de respuesta
    
    #recorremos la imagen original
    for j in range(img.shape[1]):
        for i in range(img.shape[0]):
            ans[factor*i, factor*j] = img[i,j]
    return ans
```

## Interpolación vecino más cercano

```python
def vecino_cercano(img,factor=2):
    #Control de color o blanco y negro
    if len(img.shape) is 2:
        ans = np.zeros([img.shape[0] * factor, img.shape[1] * factor], dtype=np.uint8)
    else:
        ans = np.zeros([img.shape[0] * factor, img.shape[1] * factor, 3], dtype=np.uint8)

    #recorremos la imagen original
    for j in range(img.shape[1]):
        for i in range(img.shape[0]):
            #recorrer los recuadros vacios y llenarlos con el mismo valor
            for p in range(factor):
                for q in range(factor):
                    ans[factor*i+q, factor*j+p] = img[i,j]
    return ans
```

## Interpolación Bilineal y por Bsplines
En la interpolacion bilineal tratamos cada eje, $x$ e $y$ independientemente para hacer la interpolacion que se encuentra en los puntos intermedios. Como muestra la figura.

![image]( https://upload.wikimedia.org/wikipedia/commons/thumb/9/91/Bilinear_interpolation_visualisation.svg/170px-Bilinear_interpolation_visualisation.svg.png)

```python
def interpola_Bsplin(img, factor=2, grado=3, smooth=0):
    img = img.astype(np.float32)
    
    #Control de color o blanco y negro
    if len(img.shape) is 2:
        ans = np.zeros([img.shape[0] * factor, img.shape[1] * factor], dtype=np.float32) #final
        aux = np.zeros([img.shape[0] * factor, img.shape[1] ], dtype=np.float64) #Paso intermedio
    else:
        ans = np.zeros([img.shape[0] * factor, img.shape[1] * factor, 3], dtype=np.float32)
        aux = np.zeros([img.shape[0] * factor, img.shape[1], 3 ], dtype=np.float64) #Paso intermedio

    #Calcular los puntos intermedios de las columnas
    i = 0
    while i < img.shape[1]: #recorremos la imagen original
        base = img[:,i] #La base de los puntos conocidos. (y)
        ind = np.arange(base.shape[0]) #indice de los puntos conocidos. (x)
        ind2 = np.arange(base.shape[0], step=1/float(factor)) #indice para los puntos destino. (x2)
        
        # Obtener los coeficientes de la funcion Bspline
        if len(img.shape) is 2: #imagenes a BN
            coef = interpolate.splrep(ind,base,k=grado, s=smooth, per=True) # Obtener los coeficientes de la funcion Bspline
            aux[:,i] = interpolate.splev(ind2,coef); #Guardamos los vectores en la variable auxiliar (z)
        else:        #Imagenes a color
            for channel in range(3):
                coef = interpolate.splrep(ind,base[:,channel],k=grado, s=smooth, per=True) #Coeficientes de la funcion Bspline
                aux[:,i,channel] = interpolate.splev(ind2,coef); #Guardamos los vectores en la variable auxiliar (z)
        i+=1
    
    aux = aux * np.max(img) / np.max(aux)
    
    #Calcular los puntos intermedios de las filas en base a la matriz auxiliar
    j = 0
    while j < aux.shape[0]:
        base = aux[j,:] #La base. los puntos conocidos. (y)
        ind = np.arange(base.shape[0]) #indice de los puntos conocidos. (x)
        ind2 = np.arange(base.shape[0], step=1/float(factor)) #indice para los puntos destino. (x2)
        
        if len(img.shape) is 2: #imagenes a BN
            coef = interpolate.splrep(ind,base,k=grado, s=smooth, per=True) # Obtener los coeficientes de la funcion Bspline
            ans[j,:] = interpolate.splev(ind2,coef); #Guardamos los vectores en la variable auxiliar (z)
        else:        #Imagenes a color
            for channel in range(3):
                coef = interpolate.splrep(ind,base[:,channel],k=grado, s=smooth, per=True) #Coeficientes de la funcion Bspline
                ans[j,:,channel] = interpolate.splev(ind2,coef); #Guardamos los vectores en la variable auxiliar (z)
        j+=1
    ans = ans * np.max(img) / np.max(ans)    
    return ans.astype(np.uint8)
```

```python
img = cv2.imread('../peppers_color.tif',1)
factor = 4

print "Dimenciones"
print "Img:", np.array(img.shape[0:2])
print "Ans:", np.array(img.shape[0:2]) * factor

sin = sin_interpolar(img, factor)
vecino = vecino_cercano(img,factor);
lineal = interpola_Bsplin(img,factor, grado=1);
cubico = interpola_Bsplin(img,factor);
grd5 = interpola_Bsplin(img,factor,grado=5 );

cv2.imwrite('../ANS/vecino.png',vecino)
cv2.imwrite('../ANS/lineal.png', lineal)
cv2.imwrite('../ANS/cibico.png', cubico)
cv2.imwrite('../ANS/grd5.png', grd5)
cv2.imwrite('../ANS/sin.png',sin)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
sin = cv2.cvtColor(sin, cv2.COLOR_BGR2RGB)
vecino = cv2.cvtColor(vecino, cv2.COLOR_BGR2RGB)
lineal = cv2.cvtColor(lineal, cv2.COLOR_BGR2RGB)
cubico = cv2.cvtColor(cubico, cv2.COLOR_BGR2RGB)
grd5 = cv2.cvtColor(grd5, cv2.COLOR_BGR2RGB)

f, sub = plt.subplots(1,2)
f.set_size_inches(12,5)

sub[0].imshow(sin)
sub[1].imshow(img)

plt.show()
```

```python
x_init = 140 * factor
y_init = 30 * factor
dim_lado = 250

f, sub = plt.subplots(2,2,sharex=True, sharey=True)
f.set_size_inches(12,12)

sub[0,0].imshow( vecino[y_init:y_init+dim_lado, x_init:x_init+dim_lado] ); sub[0,0].set_title("Vecino mas cercano")
sub[0,1].imshow( lineal[y_init:y_init+dim_lado, x_init:x_init+dim_lado] ); sub[0,1].set_title("Linear")
sub[1,0].imshow( cubico[y_init:y_init+dim_lado, x_init:x_init+dim_lado] ); sub[1,0].set_title("Bspline cubico")
sub[1,1].imshow( grd5[y_init:y_init+dim_lado, x_init:x_init+dim_lado] ); sub[1,1].set_title("Bspline Grado 5")

plt.show()
```
